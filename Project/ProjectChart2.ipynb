# Downloading and Editing Data from World Bank API:

```python
# Begin by installing the packages that we need got 
import numpy as np
import pandas as pd
```

```python
# Download World Bank API Packages
!pip install world-bank-data
import world_bank_data as wb
```

```python
# Select the Data that we want

urbanpop = pd.DataFrame(wb.get_series('SP.URB.TOTL.IN.ZS'))
gdpcapita = pd.DataFrame(wb.get_series('NY.GDP.PCAP.CD'))
population = pd.DataFrame(wb.get_series('SP.POP.TOTL'))
```

```python
urbanpop.to_csv('urbanpop.csv')
```

```python
# Sort and filter the urban population data to what we want to display

urbanpop = pd.read_csv('urbanpop.csv')
urbanpop = urbanpop.drop('Series', axis=1)
urbanpop = urbanpop[urbanpop.Year > 2010]
urbanpop = urbanpop.dropna()
urbanpop = urbanpop.sort_values(by=['Country', 'Year'])
urbanpop = urbanpop.drop_duplicates(subset=['Country'], keep='last')
urbanpop = urbanpop.set_index('Country')
urbanpop.columns = ["Year", "Urban population"]
urbanpop= urbanpop.drop('Year', axis=1)
```

```python
# Check that this is what we want

urbanpop
```

```python
gdpcapita.to_csv('gdpcapita.csv')
```

```python
# Sort and filter GDP capita data to what we want to display
```

```python
gdpcapita = pd.read_csv('gdpcapita.csv')
gdpcapita = gdpcapita.drop('Series', axis=1)
gdpcapita = gdpcapita[gdpcapita.Year > 2010]
gdpcapita = gdpcapita.dropna()
gdpcapita = gdpcapita.sort_values(by=['Country', 'Year'])
gdpcapita = gdpcapita.drop_duplicates(subset=['Country'], keep='last')
gdpcapita = gdpcapita.set_index('Country')
gdpcapita.columns = ["Year", "GDP per Capita"]
gdpcapita= gdpcapita.drop('Year', axis=1)
```

```python
# Check this is what we want

gdpcapita
```

```python
# Now we have the data that we want we can merge the datasets 

urbanpop_gdpcapita = pd.merge(urbanpop, gdpcapita, how='left', on='Country')
```

```python
# Check this is what we want

urbanpop_gdpcapita
```

```python
population.to_csv('population.csv')
```

```python
population = pd.read_csv('population.csv')
population = population.drop('Series', axis=1)
population = population[population.Year > 2010]
population = population.dropna()
population = population.sort_values(by=['Country', 'Year'])
population = population.drop_duplicates(subset=['Country'], keep='last')
population = population.set_index('Country')
population.columns = ["Year", "Population"]
population= population.drop('Year', axis=1)

population
```
